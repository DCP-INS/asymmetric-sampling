```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.gridspec as gridspec
from matplotlib.ticker import FormatStrFormatter
from matplotlib import ticker
```

```python
# loading ITPC spectrums from left aac 
aac_left = np.load("/home/jeremy/Desktop/itpc_spectrums_pure_tones_aac_left.npy")

# loading ITPC spectrums from right aac 
aac_right = np.load("/home/jeremy/Desktop/itpc_spectrums_pure_tones_aac_right.npy")

# loading similarity indexes for left and right aac
yl,xl,yr,xr = np.load("/home/jeremy/Desktop/similarity_index_pure_tone.npy")
```

```python
freqs= np.logspace(*np.log10([2, 250]), num=200)

fig4 = plt.figure(figsize=(6, 8))

#####################################################################################

ax0 = plt.subplot2grid((2, 2), (0, 0))
for i in aac_left:
    ax0.plot(freqs,i,color="darkkhaki", linewidth=0.5)
ax0.plot(freqs,np.mean(aac_left,0),color="darkkhaki", linewidth=6)
ax0.spines['top'].set_visible(False)
ax0.spines['right'].set_visible(False)
ax0.set_xscale("log")
ax0.set_xlim(2,250)
ax0.set_ylim(0,1)
ax0.set_yticks(np.round(np.linspace(0, 0.8,5), 2))
ax0.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax0.axes.get_xaxis().set_visible(False)
ax0.text(50,0.8,"AAC Left", color = 'k',  fontweight='bold', fontsize = 15)
ax0.set_ylabel('ITPC (z)',rotation = 90, fontsize = 20)
ax0.text(0.95,0.9, "A", fontsize = 20, weight = 'bold')
###########################################

ax1 = plt.subplot2grid((2, 2), (0, 1))
for i in aac_right:
    ax1.plot(freqs,i,color="royalblue", linewidth=0.5)
ax1.plot(freqs,np.mean(aac_right,0),color="royalblue", linewidth=6)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.set_xscale("log")
ax1.set_xlim(2,250)
ax1.set_ylim(0,1)
ax1.set_yticks(np.round(np.linspace(0.2, 0.8,4), 2))
ax1.xaxis.set_major_formatter(FormatStrFormatter("%.0f"))
ax1.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax1.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
ax1.xaxis.set_minor_formatter(ticker.NullFormatter())
ax1.xaxis.set_minor_locator(ticker.NullLocator())
n_values = 7
tick_vals = np.array([2,5,10,25,50,100,250])
tick_vals = tick_vals.astype(int)
ax1.set_xticks(tick_vals)
ax1.set_xlabel('Frequency (Hz)', fontsize = 15)
ax1.text(40,0.8,"AAC Right", color = 'k',  fontweight='bold', fontsize = 15)
ax1.text(0.75,0.9, "B", fontsize = 20, weight = 'bold')

################################################

ax3 = plt.subplot2grid((2, 2), (1, 0))
ax3.plot(freqs,aac_left[-1],color="darkkhaki",label ='AAC left hemisphere', linewidth=3)
ax3.plot(freqs,aac_right[-1],color="royalblue", label = 'AAC right hemisphere', linewidth=3)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.set_xscale("log")
ax3.set_xlim(2,250)
ax3.set_ylim(0,1)
ax3.set_yticks(np.round(np.linspace(0, 0.8,5), 2))
ax3.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax3.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax3.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
ax3.xaxis.set_minor_formatter(ticker.NullFormatter())
ax3.xaxis.set_minor_locator(ticker.NullLocator())
n_values = 7
tick_vals = np.array([2,5,10,25,50,100,250])
tick_vals = tick_vals.astype(int)
ax3.set_xticks(tick_vals)
ax3.set_xlabel('Frequency (Hz)', fontsize = 20)
ax3.set_ylabel('ITPC (z)',rotation = 90, fontsize = 20)
ax3.legend()
ax3.set_title( "bilaterally implanted patient #93", fontsize=15)
ax3.text(0.95,0.9, "C", fontsize = 20, weight = 'bold')


###################################################################
ax2 = plt.subplot2grid((2, 2), (1, 1))
ax2.scatter(yl,xl, color = 'darkkhaki',label = 'left implanted patients', linewidth='3', marker = '*', s = 100)
ax2.scatter(yr,xr, color = 'royalblue',label = 'right implanted patients', linewidth='5',marker = '+', s = 100)

ax2.plot(np.arange(0,0.55,0.01),np.arange(0,0.55,0.01), color = 'k', linestyle = ':')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax2.set_xlabel("Similarity with right AAC ITPC pattern", fontsize = 14)
ax2.set_ylabel("Similarity with left AAC ITPC pattern", fontsize = 14)
ax2.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
ax2.xaxis.set_minor_formatter(ticker.NullFormatter())
ax2.xaxis.set_minor_locator(ticker.NullLocator())
n_values = 5
tick_vals = np.array([0.2,0.4,0.6,0.8,1])
ax2.set_xticks(tick_vals)
ax2.get_yaxis().set_major_formatter(ticker.ScalarFormatter())
ax2.yaxis.set_minor_formatter(ticker.NullFormatter())
ax2.yaxis.set_minor_locator(ticker.NullLocator())
n_values = 5
tick_vals = np.array([0.2,0.4,0.6,0.8,1])
ax2.set_yticks(tick_vals)
plt.legend(loc = 'lower right');
ax2.set_xlim(0.1,1.05)
ax2.set_ylim(0.1,0.6)
ax2.yaxis.set_label_coords(-0.1,0.5)
ax2.text(-0.095,0.55, "D", fontsize = 20, weight = 'bold')
plt.subplots_adjust(left=0.1, right=2, top=0.9, bottom=0.1)
```
